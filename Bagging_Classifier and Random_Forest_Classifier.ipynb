## Bagging Classfier with train_test_split

```python
data = pd.read_csv("diabeties.csv")
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
X = data.iloc[:,0:8]
```

```python
y = data.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=356)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
BC = BaggingClassifier(n_estimators=5)
BC.fit(X_train,y_train)
```

```python
Y_train_pred = BC.predict(X_train)
```

```python
accuracy_score(y_train,Y_train_pred)
```

```python
y_test_pred = BC.predict(X_test)
```

```python
accuracy_score(y_test,y_test_pred)
```

## Random Forest Classfier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RC = RandomForestClassifier()
RC.fit(X_train,y_train)
```

```python
Y_train_pred1 = RC.predict(X_train)
```

```python
accuracy_score(y_train,Y_train_pred1)
```

```python
y_test_pred1 = RC.predict(X_test)
```

```python
accuracy_score(y_test,y_test_pred1)
```
